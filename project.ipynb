```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
from scipy.io import wavfile
```

```python
sr, data = wavfile.read("capped.wav")
```

```python
duration = len(data) / sr 
time_array = np.linspace(0, duration, len(data)) 
```

```python
time_array
```

```python
if data.ndim > 1:
        data = data[:, 0]

 # Perform FFT
fft_result = np.fft.fft(data)

    # Calculate frequencies corresponding to the FFT result
frequencies = np.fft.fftfreq(len(data), d=1/sr)

    
positive_frequencies = frequencies[:len(data)//2]
magnitude_spectrum = np.abs(fft_result[:len(data)//2])
```

```python

    # Plot waveform
plt.figure(figsize=(10, 4))
plt.plot(time_array, data)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Audio Waveform of closed resonator")
plt.show()

    # Plot frequency spectrum
plt.figure(figsize=(10, 4,))
plt.plot(positive_frequencies, magnitude_spectrum)
plt.xlim(40, 1100)  
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")
plt.title("Frequency Spectrum of closed resonator")
plt.show()
```

```python
plt.specgram(data)
plt.figure(figsize=(10, 4,))
```

```python
plt.figure(figsize=(10, 4,))
plt.specgram(data)

plt.xlim(0, 1e6)  # Set x-axis limits
plt.ylim(0, 0.2) # Set y-axis limits
```
